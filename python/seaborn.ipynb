```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
sns.set_style('ticks')
```

```python
dados = sns.load_dataset('tips')
```

```python
dados.head()
```

# Gráficos de distribuição

## Distplot - Distribuição com KDE

```python
sns.set_style('darkgrid')
sns.distplot(dados.total_bill)
```

## Jointplot - mostra a distribuição entre duas variáveis

```python
sns.jointplot(x=dados.total_bill, y=dados.tip, data=dados)
```

## Pairplot - Faz jointplots de todas as variaveis numéricas

```python
sns.pairplot(dados, hue='sex')
```

```python
#usando outro dado com cor diferente
sns.set_style('white')
sns.pairplot(dados, hue='sex')
```

# Plots Categóricos

```python
sns.heatmap(dados.corr(), cmap='Blues',)
```

# Plots Categoricos

## Barplots

```python
sns.barplot(x='sex', y='total_bill', data=dados)
```

## Countplot

```python
sns.countplot(x='sex', data=dados)
```

## Boxplot

```python
plt.figure(figsize=(12,12))
sns.set_style('white')
sns.boxplot(data=dados)
```

```python
sns.boxplot(x='day', y='total_bill', data=dados)
```

```python
sns.boxplot( data=dados, orient='h')
```

## Violinplot - mostra a distribuição estimadas dos dados baseados no KDE

```python
sns.violinplot(x='day', y='tip', data=dados)
```

```python
# mostra as distribuições com mais de uma variável
sns.violinplot(x='day', y='tip', data=dados, hue='sex')
```

```python
# como o que aparece do lado esquerdo é igual ao direito, é possivel juntar os dois
sns.violinplot(x='day', y='tip', data=dados, hue='sex',split=True)
```

# Plots de regressão

```python
sns.lmplot(x='total_bill', y='tip', data=dados)
```

```python
sns.lmplot(x='total_bill', y='tip', data=dados, hue='sex', markers=['o','v'])
```

```python
# para separar em dois ou mais graficos usar o parametro col ou row
sns.lmplot(x='total_bill', y='tip', data=dados, col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=dados, row='sex')
```
