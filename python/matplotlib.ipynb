# Matplotlib - Exerc√≠cios

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
y = x **2
```

```python
# usando como programa√ß√£o sequencial
plt.plot(x,y,'r--')
plt.xlabel('Escala de X')
plt.ylabel('Escala Y')
plt.title('Titulo')
```

### Multiplos graficos com subplot
Essa fun√ß√£o cria uma matriz de plots, recebe a linha, coluna e qual grafico que ser√° usado

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b--')
plt.subplot(1,2,2)
plt.plot(y,x,'r--')
```

### Criando um figure manualmente

```python
# figure cria uma tela de desenho para os gr√°ficos
# add_axes recebe margem esquerda, baixo, largura e altura
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.9,0.9])
axes.plot(x,y)
axes.set_xlabel('Eixo X')
axes.set_ylabel('Eixo Y')
axes.set_title('Gr√°fico 1')

axes2 = fig.add_axes([0.15,0.5,0.5,0.3])
axes2.plot(x,y)
axes2.set_xlabel('Eixo X')
axes2.set_ylabel('Eixo Y')
axes2.set_title('Gr√°fico 2')
```

### Usando subplots
retorna um figure e uma lista de eixos

```python
fig, axes = plt.subplots(1,2)
```

```python
axes[0].plot(x,y)
axes[1].plot(y,x)
fig
```
